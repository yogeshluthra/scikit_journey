#### ROC code understanding

```python
import numpy as np
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
y_true=np.array([1, 1, 2, 2])
y_pred=np.array([0.1, 0.4, 0.35, 0.8])
(fpr, tpr, threshold) = roc_curve(y_true, y_pred, pos_label=2)
print fpr
print tpr
print threshold
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.show()
```
